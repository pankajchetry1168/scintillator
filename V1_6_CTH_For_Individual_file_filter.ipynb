```python
#get max peak [first step]
#200 and 300 signal peak
#max peak histogram
#noise removal average the noise[last]
#----------------------------------------------------------------------------------------------------------------------#
#*Main test program to work on*
import matplotlib.pyplot as plt
import numpy as np
##-----------------------------------------------------------------------------------
def fn_int(chnl,clr,chnum): # channel=chnl; clr=colour; chnum=channel number ##Function for integration/summation.
    period=512
    temp1=0
    start=230 # rise of signal
    rang=120  # duration of peak
    intcha1=[]
    for intg in range(0,512,1):
        temp1=np.array(chnl[start:start+rang]).sum()
        intcha1.append(temp1)
        start=start+period
    plt.title('Integration of the peak region of signal of channel %i' %chnum)        
    plt.figure(1)
    plt.plot(intcha1,clr)#using clr for colour eg 'yellow'
    plt.show()

##----------------------------------------------------------------------------------------------
# function to enter values into each each channel
def fn_input(wavefrm,sign,err,start1): #wavefrm=waveformData[], sign= positive or negative signal err= error condidered start1=
    channela1_=[]                      #start1=location to start "jumping" in the step of 4 as per the storage of the signal
    for row in range(start1,len(wavefrm),4):     # Print(pnt,waveformData[pnt])
        thrsld=sign*err*min(wavefrm[row])       #setting threshold value with "2%" error and '-' sign for negative signal
        for col in wavefrm[row]:
            col=sign*col                           #waveform was negative to make it positive
            if(col>thrsld):                      #for proper signal 2% error is taken as threshold w.r.t max signal at that trigger[512]
                channela1_.append(col)
            else: 
                channela1_.append(0)
    return channela1_

##------------------------------------------------------------------------------------------------
def fn_hist(channela1_):
    hchannela1_=[]
    period=512
    start=0
    for i in range(0,len(channela1_),period):        # program for histrogram each channel
        maxx=max(channela1_[start:start+period])
        hchannela1_.append(maxx)
        start=start+period
    return hchannela1_
#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
#"Averaging of signal of individual channel(average of all signal in channel)"

def fn_comb(channela1_,clr,chnm):#channela1_= channel array variable,clr= colour of fig,chnm= channel number
    count=0
    comb1=[0]*v1    #v1=512
    for i in channela1_:
        if(count==v1):
            count=0
        comb1[count]=comb1[count]+i
        count+=1
    for i in range(512):
        comb1[i]=comb1[i]/1000 #averaging the signal of the channel(1000 trigger in a channel)
    plt.title('Average of signal of channel %i (Averaging of all signal of the channel)' %chnm) 
    plt.figure(1)
    plt.plot(comb1,clr)
    plt.show()

#------------------------------------------------------------------------------------------------
###---------------------------------------------------------------------------------------------------------------
###http://databookuw.com/ Chapter 2-Denoising
def fn_denoising(channela1__):
    #n=len(channela1)
    n=512000
    dt=1
    t=range(0,512000)
    f=channela1__[0:512000]
    fhat = np.fft.fft(f,n)
    PSD = fhat*np.conj(fhat)/n
    freq =(1/(dt*n))*np.arange(n)
    #L=np.arange(1,np.floor(n),dtype='int')
    L=np.arange(512)
    ## Use the PSD to filter out noise
    indices = PSD > 100       # Find all freqs with large power
    PSDclean = PSD * indices  # Zero out all others
    fhat = indices * fhat     # Zero out small Fourier coeffs. in Y #I didn't understand this part
    ffilt = np.fft.ifft(fhat) # Inverse FFT for filtered time signal
    return ffilt
    
#     fig,axs = plt.subplots(2,1)

#     plt.sca(axs[0])
#     plt.plot(t,f,color='r',LineWidth=1.5,label='Noisy')
#     #plt.plot(t,f_clean,color='k',LineWidth=2,label='Clean')
#     plt.xlim(t[0],t[-1])
#     plt.legend()

#     plt.sca(axs[1])
#     #plt.plot(t,f_clean,color='k',LineWidth=1.5,label='Clean')
#     plt.plot(t,ffilt,color='b',LineWidth=2,label='Filtered')
#     plt.xlim(t[0],t[-1])
#     plt.legend()
#     ##--------------------------------------------------------------------
#     ##Important Section for PSD visual comparision
#     # plt.sca(axs[2])
#     # plt.plot(freq[L],PSD[L],color='r',LineWidth=2,label='Noisy')
#     # plt.plot(freq[L],PSDclean[L],color='b',LineWidth=1.5,label='Filtered')
#     # plt.xlim(freq[L[0]],freq[L[-1]])
#     ##----------------------------------------------------------------------
#     plt.legend()
#     plt.show()
###---------------------------------------------------------------------------------------------------------------

file="test_data/Test80.0V_Thre6_1.npy"
filedata=np.array(np.load(file,allow_pickle=True))
waveformData=filedata[6]                #making 2D [nX6] into 1D[1X4000]. So, n should be integer but its comming around 666.667.
print(filedata[0])                      #making 2D [nX6] into 1D[1X4000]. So, n should be integer but its comming around 666.667.
print(filedata[1])
print(filedata[2])                      #need to understand
print(filedata[3])
print(filedata[4])
print(filedata[5])
#print(filedata[6])
NumberOfWaveforms=len(waveformData)
#print(len(waveformData[0]))  # for number of columns 512
#print(len(waveformData))     # number of rows       4000 (each channel 1000 rows)
#print(len(waveformData[]))

# print(NumberOfWaveforms)=4000
# the first 6 have some hadder info. 7th is waveform data. 7th have 4000 data.
# print header inside print.

channela1=[]
channelb1=[]
channelc1=[]
channeld1=[]
hchannela1=[]
hchannelb1=[]
hchannelc1=[]
hchanneld1=[]
thrsld=0.0     # initilizing the threshold
v1=512         # number of values in a signal i.e 512
err=0.00       # percentage of error 0 i.e negative signal removed
#data_points=2048                            # The size of one channel is 1000. So to reduce the length of the each channel into 
                                             # Half (512 x-axis) => 2048/4channels.

##-----------------------------------------------------------------------------------------------------------------------------
##In put values to channel 1
sign=-1 #for negative and positive peak
start1=0 #starting point for jumping
channela1=fn_input(waveformData,sign,err,start1)
##------------------------------------------------------------------------------------------------------------------------------

##-----------------------------------------------------------------------------------------------------------------------------
##In put values to channel 2
sign=1 #for negative and positive peak
start1=1 #starting point for jumping
channelb1=fn_input(waveformData,sign,err,start1)
##------------------------------------------------------------------------------------------------------------------------------

##-----------------------------------------------------------------------------------------------------------------------------
##In put values to channel 3
sign=-1 #for negative and positive peak
start1=2 #starting point for jumping
channelc1=fn_input(waveformData,sign,err,start1)
##------------------------------------------------------------------------------------------------------------------------------
    
##-----------------------------------------------------------------------------------------------------------------------------
##In put values to channel 4
sign=-1 #for negative and positive peak
start1=3 #starting point for jumping
channeld1=fn_input(waveformData,sign,err,start1)
##------------------------------------------------------------------------------------------------------------------------------

##------------------------------------------------------------------------------------------------
# program for histrogram channel 1

hchannela1=fn_hist(channela1)
#------------------------------------------------------------------------------------------------

##------------------------------------------------------------------------------------------------
# program for histrogram channel 2
hchannelb1=fn_hist(channelb1)
#------------------------------------------------------------------------------------------------

##------------------------------------------------------------------------------------------------
# program for histrogram channel 3
hchannelc1=fn_hist(channelc1)
#------------------------------------------------------------------------------------------------

##------------------------------------------------------------------------------------------------
# program for histrogram channel 3
hchanneld1=fn_hist(channeld1)
#------------------------------------------------------------------------------------------------

#------------------------------------------------------------------------------------------------
#"Combination of signal of channel 1(Overlap of all signal in channel 1)"
##Sub part for "red curve from each actual waveform, not the histogram" 

fn_comb(channela1,'yellow',1)
#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
#"Combination of signal of channel 1(Overlap of all signal in channel 2)"
##Sub part for "red curve from each actual waveform, not the histogram" 

fn_comb(channelb1,'red',2)
#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
#"Combination of signal of channel 3(Overlap of all signal in channel 3)"
##Sub part for "red curve from each actual waveform, not the histogram" 

fn_comb(channelc1,'black',3)
#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
#"Combination of signal of channel 4(Overlap of all signal in channel 4)"
##Sub part for "red curve from each actual waveform, not the histogram" 

fn_comb(channeld1,'green',4)
#------------------------------------------------------------------------------------------------
##Integration of the peak portion of channel 1
# Area = simps((channela1)[StartIndex:EndIndex],TimeArray[StartIndex:EndIndex]) #Sam code

fn_int(channela1,'yellow',1)

#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
##Integration of the peak portion of channel 2
# Area = simps((channela1)[StartIndex:EndIndex],TimeArray[StartIndex:EndIndex]) #Sam code

fn_int(channelb1,'red',2)

#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
##Integration of the peak portion of channel 3

fn_int(channelc1,'black',3)

#------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------
##Integration of the peak portion of channel 4

fn_int(channeld1,'green',4)


#fig.suptitle("Scintiallator Signal[First 512]")
fig, axs1=plt.subplots(2,2)
axs1[0,0].plot(channela1[0:v1],'yellow')
axs1[0,0].set_title("Channel 1")
axs1[0,1].plot(channelb1[0:v1],'red')
axs1[0,1].set_title("Channel 2")
axs1[1,0].plot(channelc1[0:v1],'black')
axs1[1,0].set_title("Channel 3")
axs1[1,1].plot(channeld1[0:v1],'green')
axs1[1,1].set_title("Channel 4")
fig.tight_layout()
#axs[1, 0].set_title("shares x with main")

#fig.suptitle("Scintiallator Signal Maximum Value of all channels\n")
fig, axs1=plt.subplots(2,2)
axs1[0,0].hist(hchannela1,bins='auto')
axs1[0,0].set_title("Channel 1")
axs1[0,1].hist(hchannelb1,bins='auto')
axs1[0,1].set_title("Channel 2")
axs1[1,0].hist(hchannelc1,bins='auto')
axs1[1,0].set_title("Channel 3")
axs1[1,1].hist(hchanneld1,bins='auto')
axs1[1,1].set_title("Channel 4")
fig.tight_layout()
plt.show() # c and d trigger

#------------------------------------------------------------------------------------------------
#fft
channela1=fn_denoising(channela1)
channelb1=fn_denoising(channelb1)
channelc1=fn_denoising(channelc1)
channeld1=fn_denoising(channeld1)
#-------------------------------------------------------------------------------------------------
period=512
start=0
for i in range(50):
    fig, axs1=plt.subplots(2,2)
    axs1[0,0].plot(channela1[(start):(start+period)],'yellow',label='Channel 1')
    #axs1[0,0].set_title("Channel 1")
    axs1[0,1].plot(channelb1[(start):(start+period)],'red',label='Channel 2')
    #axs1[0,1].set_title("Channel 2")
    axs1[1,0].plot(channelc1[(start):(start+period)],'black',label='Channel 3')
    #axs1[1,0].set_title("Channel 3")
    axs1[1,1].plot(channeld1[(start):(start+period)],'green',label='Channel 4')
    #axs1[1,1].set_title("Channel 4")
    start=start+period
    plt.legend()
    plt.show()
```
